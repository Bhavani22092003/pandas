### 2. Aggregate DataFrame df by extracting the month and year from the 'Date' column. Calculate the collective revenue for each unique combination of month and year.


```python
data['Date'].unique() 
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data['Month'] = data['Date'].dt.month_name() ## it prints month names as per month numbers 
data['Year'] = data['Date'].dt.year ## it prints year ,in a date ,by using .dt.year we can extract the year .
data[['Date','Month','Year']]
```

```python
data.groupby(['Year','Month'])['Revenue'].sum() 
```

### 3. Group the DataFrame df by the 'Date' column and find the date with the highest total revenue and the date with the highest total quantity sold.

```python
max_revenue_date,max_total_revenue = data.loc[data['Revenue'].idxmax(), 'Date'],data['Revenue'].max()
```

```python
print(f'The Highest quantity revenue is  : {max_total_revenue} ,  sold in the Date of  : { max_revenue_date }')
```

```python
max_revenue_Quantity,max_revenue_quantity = data.loc[data['Quantity'].idxmax(), 'Date'],data['Quantity'].max()
```

```python
print(f'The Highest quantity is : {max_revenue_quantity } ,  sold in the Date of  : { max_revenue_Quantity }')
```

### 5. Group the DataFrame df by the 'Product' column and add a new column 'Revenue Percentage' that shows the percentage of each product's revenue relative to the total revenue for that product.


```python
pd.read_csv(r'C:\Users\Bhavani\Downloads\groupbyExcercise - groupbyExcercise.csv')
```

```python
data['Total Revenue'] = data.groupby('Product')['Revenue'].transform('sum')
```

 > transform() 
* Is used to perform operations on groups of data and then broadcast the results back to the original DataFrame. 

```python
data
```

```python
data['Revenue Percentage']=(data['Revenue'] / data['Total Revenue']) * 100
```
